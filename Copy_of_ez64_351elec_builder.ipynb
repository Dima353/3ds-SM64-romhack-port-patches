### Download and extract the content in the roms/ports folder of your RG351. You should have a new entry "sm64" appearing in your "Ports" section (If it doesn't appear, try to restart emulationstation. If you don't know how to do it, just reboot your console). Launch it and enjoy !

Mapping :

* N64 A : RG351 A
* N64 B : RG351 B
* N64 START : RG351 START
* N64 Z : RG351 L
* N64 R : RG351 R
* N64 BUTTONS : RG351 left stick

You can quit the game by pressing SELECT + START simultaneously
